# Tokenizer

<div dir=rtl>
در مرحله اول جملات موجود در تمامی توییت ها را با استفاده از کلاس Tokenizer و متد tokenize آن توکنایز می کنیم و در آرایه all_tokens ذخیره می کنیم.

# Normalizer

<div dir=rtl>
در مرحله دوم توکن های بدست آمده در مرحله قبل را با استفاده از کلاس Normalizer و متد normalize آن نرمالایز می کنیم و در آرایه all_normalized_tokens ذخیره می کنیم.

این نرمال سازی شامل موارد زیر است:
* حذف علایم نگارشی
* تبدیل حروف بزرگ به کوچک
* تبدیل اعداد به کلمات
* حذف حروف مربوط به رشته
* حل مشکلات نگارشی (به مقدار محدود)

یکی از مواردی که می توان در نرمال سازی استفاده کرد یافتن نزدیک کلمه با استفاده از فاصله levenshtein است، اما چون این مورد زمان زیادی می گیرد از آن صرف نظر شد و حل مشکلات نگارشی به مقدار محدود به تنهایی استفاده شد.
</div>

```python
from Normalizer import Normalizer

all_normalized_tokens = []
normalizer = Normalizer()

for tokens in tqdm.tqdm_notebook(all_tokens):
    all_normalized_tokens.append(normalizer.normalize(tokens))
```

<div dir=rtl>
با توجه به اینکه اجرای سلول الا حدود 15 دقیقه زمان می برد، برای صرفه جویی در زمان برای تست مراحل بعد آرایه نهایی بدست آمده با استفاده از کتابخانه pickle را ذخیره میکنیم.
</div>

```python
# write in a file for latter use
import pickle

with open('all_normalized_tokens.ob', 'wb') as fp:
    pickle.dump(all_normalized_tokens, fp)
```

<div dir=rtl>
پس از ذخیره آرایه بدست آمده می توان هر زمان که بخواهیم آن را لود کنیم.
</div>

```python
# load from file quickly
import pickle

with open ('all_normalized_tokens.ob', 'rb') as fp:
    all_normalized_tokens = pickle.load(fp)
```

# Stopword removing

<div dir=rtl>
در مرحله سوم، با استفاده از کلاس StopwordRemover و صدا کردن متد remove_stopwords آن، تمامی stopword ها را از آرایه شامل توکن های نرمالایز شده حذف می کنیم.  برای تشخیص stopword بودن یا نبودن از یک فایل شامل stopwordهای ترکی استفاده می کنیم.
</div>

```python
from StopwordRemover import StopwordRemover

swr = StopwordRemover()

for i, tokens in tqdm.tqdm_notebook(enumerate(all_normalized_tokens)):
     all_normalized_tokens[i] = swr.remove_stopwords(tokens)
```

# Stemmer & Lemmatizer

<div dir=rtl>

</div>

```python
from Stemmer import Stemmer

stemmer = Stemmer()


for i, tokens in tqdm.notebook.tqdm(enumerate(all_normalized_tokens)):
    stems = []
    for token in tokens:
        stems.append(stemmer.stemmer(token))
    all_normalized_tokens[i] = stems
```

## Before Preprocessing

```python
language = "tr"
max_ngram_size = 3
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 20

custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)
keywords = custom_kw_extractor.extract_keywords(" ".join(tweets))

for kw in keywords:
    print(kw)
```

## After Preprocessing

```python
language = "tr"
max_ngram_size = 2
deduplication_thresold = 0.9
deduplication_algo = 'seqm'
windowSize = 1
numOfKeywords = 20

custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_thresold, dedupFunc=deduplication_algo, windowsSize=windowSize, top=numOfKeywords, features=None)

text = " ".join([" ".join(tweet) for tweet in all_normalized_tokens])
keywords = custom_kw_extractor.extract_keywords(text)

for kw in keywords:
    print(kw)
```
